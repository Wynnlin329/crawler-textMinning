```python
options = Options()
options.add_argument("headless")

driver = Chrome('./chromedriver', chrome_options = options)
url ='https://www.instagram.com/explore/tags/%E7%9C%BC%E5%A6%9D/?hl=zh-tw'
driver.get(url)
title_list = []
count = 0
roll_page = 0
while roll_page < 1000:
    for page in range(1,4):
        js="var q=document.documentElement.scrollTop=100000"  
        driver.execute_script(js)  
        time.sleep(2)
        print(page)
    html_title = driver.page_source
    soup = BeautifulSoup(html_title,'html.parser')
    nono = soup.select('div[class="Nnq7C weEfm"] div[class="v1Nh3 kIKUG _bz0w"]')
    for num,title_all in enumerate(nono):
        article_url = 'https://www.instagram.com'+nono[num].a['href']
        
        if article_url not in title_list:
            driver_article =Chrome('./chromedriver', chrome_options = options)
            driver_article.get(article_url)
            pic_list=[]
            
            while count < 100000:
                time.sleep(2)
                try:
                    driver_article.find_elements_by_class_name('_6CZji')[0].click()

                    cc = driver_article.find_elements_by_class_name('FFVAD')
                    for x in cc:
                        #print(x.get_attribute('src').split(',')[-1][:-4])
                        pic =x.get_attribute('src').split(',')[-1][:-4]
                        if pic not in pic_list:
                            driver_pic = Chrome('./chromedriver', chrome_options = options)
                            driver_pic.get(pic)
                            screenshot_name = ".\\IG_photo_1\\eye{}.png".format(count)
                            print(screenshot_name)
                            driver_pic.save_screenshot(screenshot_name)
                            driver_pic.close()
                            pic_list.append(pic)
                        count +=1
                        
                except Exception as e:
                    title_list.append(article_url)
                    driver_article.close()
                    break
                    
    roll_page +=1

```
