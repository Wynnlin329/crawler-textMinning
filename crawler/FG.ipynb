```python
import time
import requests
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
#driver = Chrome('./chromedriver')
import requests
from bs4  import BeautifulSoup
import os
import json
from urllib import request
import json
import re
import pandas as pd
import random
```

```python
#user_agent ='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
   'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
   'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
   'Accept-Encoding': 'none',
   'Accept-Language': 'en-US,en;q=0.8',
   'Connection': 'keep-alive'}

path = './FGblog'
if not os.path.isdir(path):
    os.mkdir(path)
G =0
df = pd.DataFrame(columns =['標題','日期','作者','網址','內文'])#設定PD
```

```python
for n in range(1,200):
    url ='https://fgblog.fashionguide.com.tw/category/4/last_posts?page={}'.format(n)
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text,'html.parser')
    title =soup.select('li[class="fg-card-middle"]')
    for i,title_all in enumerate(title):
        #print(title_all.a.text)
        title_article = title_all.a.text
        title_article = title_article.replace('【','').replace('】','').replace('?','').replace('[','').replace(']','').replace('◆','').replace("/","-")
        print('')
        blog_url = title_all.a['href']
        #抓各篇內文
        url1 ='https://fgblog.fashionguide.com.tw{}'.format(blog_url)
        #print(url1)
        res1 = requests.get(url1, headers=headers)
        soup1 = BeautifulSoup(res1.text,'html.parser')
        AA = soup1.select('div[class="article-content"]')
        #print(AA[0].text.replace('\n\n',''))
        article_text =AA[0].text.replace('\n\n','')
        Time = soup1.select('div[class="post-date"]')
        author= soup1.select('div[class="profile-right"]')
        #
        DATA = []
        DATA.append(title_article.replace('\n','').replace('\r','').replace(' ',''))#標題
        DATA.append(Time[0].text)#時間
        DATA.append(author[0].a.text)#作者
        DATA.append(url1)#網址
        DATA.append(article_text.replace('\n','').replace('\r','').replace(' ',''))#內文
        df.loc[G]=DATA#DF結合DATA
        
        
        
        G+=1
        k = random.randint(5, 10)
        time.sleep(k)
        
        
```
