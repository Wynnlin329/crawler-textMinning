```python
df1=df
df1['seg']=None
df1['word_freq']=None
df1
```

```python
n=len(df)
alltext=[]
stopset=set()
stop2=['/n','']


with open('E:\\Tensorflow\\textMinning\\stopword.txt','r',encoding='utf-8') as s:
    for line in s:
        stopset.add(line.strip('\n'))
jieba.load_userdict('E:\\Tensorflow\\textMinning\\DB104-master\\userdict.txt')
for i in range(n):
    seg=''
    wf={}
    text= str(df1.loc[i]['內文'])+'\n'+str(df1.loc[i]['標題'])+'\n'
    cut=jieba.cut(text,cut_all=False)
    for j in cut:
        if j not in stopset:
            seg +=j+' '
    df1['seg'][i]=seg
    for w in seg.split(' '):
        if w not in wf:
            wf[w]=1
        else:
            wf[w]+=1
    word_list=[(k,wf[k]) for k in wf if k not in stop2 ]
    word_list.sort(key=lambda a :a[1],reverse=True)
    df1['word_freq'][i]=word_list

    
```
