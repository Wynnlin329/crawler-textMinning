# 資料缺值處理

```python
import pandas as pd
```

```python
df = pd.read_json('./crawler/FGblog.json',orient='index' )
df
```

```python
df = pd.read_excel('./FGblog/FGblog.xlsx', header=0 )
```

### 1. 統計所有欄位的 NA 值

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
missing_values_count[0:5]
```

### 2. 去除 NA 值

```python
# remove all the rows that contain a missing value
df.dropna()
```

```python
df = df.drop(columns = ['Unnamed: 0'],axis=1)
```

```python
df
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

### 3. 資料自動填補值的操作

```python
# get a small subset of the NFL dataset
subset_df = df.loc[:, 'EPA':'Season'].head()
subset_df
```

```python
# replace all NA's with 0
subset_df.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
subset_df.fillna(method = 'bfill', axis=0).fillna(0)
```

# 時間日期處理

```python
import pandas as pd

df = pd.read_csv('data/catalog.csv')
df.shape
```

```python
df.head()
```

### 1. 時間日期欄位轉換

```python
# print the first few rows of the date column
print(df['date'].head(10))
```

```python
df['date'].dtype
```

```python
# convert to datetime format
#df['date_parsed'] = pd.to_datetime(df['date'], format = "%m/%d/%y")
df['發文日期'] = pd.to_datetime(df['日期'], format = "%m/%d/%y")
```

```python
print(df['date_parsed'].head(10))
```

```python
# set index to date
df = df.set_index(df['date_parsed'])
df =df.sort_index()
df = df[df.index.notnull()]
df.head(10)
```

```python
# select row bt date
df["2007-03":"2007-05"]
```

### 2. 時間序列解析度操作

```python
# select sub df
df = df[~df.index.duplicated(keep='first')]
sub_df = df.loc[:,['population', 'distance']].head(10)
sub_df
```

```python
# resample interval to 1 day.
sub_df.resample('1D').asfreq()
```

```python
# bfill
sub_df.resample('1D').bfill()
```

# 資料分割選擇與合併

## 1. .iloc vs loc

![](http://104.236.88.249/wp-content/uploads/2016/10/Pandas-selections-and-indexing.png)

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randn(4,4), columns=['a', 'b', 'c', 'd'])
```

```python
df
```

```python
# iloc index based index slelction
df.iloc[:2,[0,2]]
```

```python
# loc label based index slelction
df.loc[:1,['a', 'c']]
```

### 兩者容易混淆之處

```python
df = pd.DataFrame(np.random.randn(4,4), columns=['a', 'b', 'c', 'd'], index=[0, 1, 8, 3])
df
```

### 選取 3, 4 列作法:

```python
# iloc
df.iloc[2:4,:]
```

```python
# loc
df.loc[2:4,:]
```

```python
# loc 正確 index 取法
df.loc[8:3,:]
```

## 2. 合併 concat

```python
df1 = pd.DataFrame(np.random.randn(4,5), columns=['a', 'b', 'c', 'd', 'e'])
df2 = pd.DataFrame(np.random.randn(6,4), columns=['g', 'y', 'k', 'f'])
print(df1.shape)
print(df2.shape)
```

```python
# 列方向合併 (上下)
merge_df = pd.concat([df1, df2], axis=0)
merge_df
```

```python
print(merge_df.shape)
```

```python
# 行方向合併 (左右)
merge_df = pd.concat([df1, df2], axis=1)
merge_df
```

```python
merge_df.shape
```
