```python
y=dict()
with open('./FGblog/FGblog_test.xlsx','r',encoding='utf-8') as j:
    x=j.read()
w=x.split(' ')
for i in w:
    if i not in y:
        y[i]=1
    else:
        y[i]+=1

        
    
```

```python
word_list=[(k,y[k]) for k in y]
word_list.sort(key=lambda a :a[1],reverse=True)
word_list
```

```python
with open(r'./wordfreq.txt','w',encoding='utf-8') as f:
    f.write(str(word_list))
```
